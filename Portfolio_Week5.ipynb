```python
import requests
import json
from google.colab import files

url_base= "https://api.stlouisfed.org/fred/series/observations?series_id={}&api_key=285d15a70cec8718abc88ae52655ef3f&file_type=json"
fredSeries = ['MKTGDPCNA646NWDB', 'NAEXKP01RUQ652S', 'CLVMNACSCAB1GQIT', 'CLVMNACSCAB1GQDE']
file_base = "Portfolio_Week5_{}.json"

for i in fredSeries:
  URL = url_base.format(i)
  print(URL)
  data = requests.get(URL).json()
  print(data)
  fileName= file_base.format(i)
  print(fileName)
  with open(fileName, 'w', encoding='utf-8') as f:
    json.dump(data,f, ensure_ascii=False, indent=4)
  files.download('{}'.format(fileName))
```
